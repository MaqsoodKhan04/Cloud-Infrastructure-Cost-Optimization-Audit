```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

# Setting seed for reproducibility
np.random.seed(42)

rows = 1000
services = ['EC2 (Compute)', 'S3 (Storage)', 'RDS (Database)', 'Lambda (Serverless)', 'CloudFront (CDN)']
departments = ['Engineering', 'Marketing', 'Data Science', 'Sales', np.nan] # Added NaN for "messy" data
environments = ['Prod', 'Dev', 'Staging']

data = []

for i in range(rows):
    date = datetime(2026, 1, 1) + timedelta(days=random.randint(0, 90))
    service = random.choice(services)
    usage_type = "InstanceUsage" if service == 'EC2 (Compute)' else "Storage"
    
# Logic to create "Zombie" resources (High cost, 0 usage)
    is_zombie = random.random() < 0.05 # 5% chance
    if is_zombie:
        cost = random.uniform(50, 200)
        usage_amount = 0.0
    else:
        cost = random.uniform(0.1, 50)
        usage_amount = random.uniform(1, 100)
        
    data.append({
        'Date': date.strftime('%Y-%m-%d'),
        'ResourceID': f"res-{random.randint(1000, 9999)}",
        'Service': service,
        'Department': random.choice(departments),
        'Environment': random.choice(environments),
        'UsageAmount': usage_amount,
        'Cost': round(cost, 2),
        'Currency': 'USD'
    })

df = pd.DataFrame(data)
df.to_csv('cloud_billing_data.csv', index=False)
print("'cloud_billing_data.csv' has been created!")
```

```python
import pandas as pd

# Loading the data
df = pd.read_csv('cloud_billing_data.csv')

#Fixing Missing Departments
#replace empty departments with "Unallocated" 
df['Department'] = df['Department'].fillna('Unallocated')

# Convert 'Date' to a proper Datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Creating a "Waste" Column
# This is a 'Flag' that marks rows where Usage is 0 but Cost is > 0
df['Is_Waste'] = (df['UsageAmount'] == 0) & (df['Cost'] > 0)

#Create a "Cost per Unit" Column
# We use np.where to avoid dividing by zero
import numpy as np
df['Unit_Cost'] = np.where(df['UsageAmount'] > 0, df['Cost'] / df['UsageAmount'], 0)

# Save the cleaned data
df.to_csv('cleaned_cloud_billing.csv', index=False)

print("Data is cleaned and 'cleaned_cloud_billing.csv' is ready!")
print(f"Total rows cleaned: {len(df)}")
print(f"Missing departments fixed: {df[df['Department'] == 'Unallocated'].shape[0]}")
```

```python
# Show top 5 resources where we are wasting the most money
waste_summary = df[df['Is_Waste'] == True].sort_values(by='Cost', ascending=False)
print("Top 5 Zombie Resources (0 Usage, High Cost):")
print(waste_summary[['ResourceID', 'Service', 'Cost']].head())
```

```python
import pandas as pd

# Loading the cleaned data
df = pd.read_csv('cleaned_cloud_billing.csv')
df['Date'] = pd.to_datetime(df['Date'])

# 1. Total Spend & Total Waste
total_spend = df['Cost'].sum()
total_waste = df[df['Is_Waste'] == True]['Cost'].sum()
waste_percentage = (total_waste / total_spend) * 100

# 2. Spend by Department
dept_analysis = df.groupby('Department').agg({
    'Cost': 'sum',
    'Is_Waste': 'sum'
}).rename(columns={'Is_Waste': 'Zombie_Count'}).sort_values(by='Cost', ascending=False)

# 3.Burn Rate
daily_trend = df.groupby('Date')['Cost'].sum()

print(f"--- GLOBAL SUMMARY ---")
print(f"Total Cloud Spend: ${total_spend:,.2f}")
print(f"Potential Savings: ${total_waste:,.2f} ({waste_percentage:.1f}% of total)")
print(f"\n--- DEPARTMENT BREAKDOWN ---")
print(dept_analysis)
```

```python
# Save the department summary for our dashboard
dept_analysis.to_csv('department_summary.csv')
# Saving the daily trend
daily_trend.to_csv('daily_burn_rate.csv')
print("Summary files generated!")
```
